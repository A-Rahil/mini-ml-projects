```python
import numpy as np
import pandas as pd
import sklearn as sk
```

```python
dataFrame = pd.read_csv(r'spam.csv', encoding='latin1')
dataFrame = dataFrame.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
dataFrame = dataFrame.rename(columns={'v1': 'Label', 'v2': 'Content'})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

count = CountVectorizer()
X = count.fit_transform(dataFrame['Content'])
Y = dataFrame['Label']


train_X, test_X, train_Y, test_Y =  train_test_split(X, Y, test_size=0.2, random_state=120)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(train_X, train_Y)
prediction = model.predict(test_X)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test_Y, prediction)
print(f"Accuracy of the model: {accuracy}")
```
