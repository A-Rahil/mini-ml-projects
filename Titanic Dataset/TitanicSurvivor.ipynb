```python
import numpy as np
import pandas as pd
```

```python
dfTrain = pd.read_csv(r'train.csv')
dfTrain.head()
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']
X = dfTrain[features]
Y = dfTrain['Survived']
```

```python
medianAge = X['Age'].median()
X['Age'] = X['Age'].fillna(medianAge)
X['Sex'] = X['Sex'].map({'male': 1, 'female': 0})
X['Embarked'] = X['Embarked'].dropna()
X = pd.get_dummies(X, columns=['Embarked'])
```

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(X, Y)
```

```python
dfTest = pd.read_csv(r'test.csv')
testX = dfTest[features]
testX['Age'] = testX['Age'].fillna(medianAge)
testX['Sex'] = testX['Sex'].map({'male': 1, 'female': 0})
testX['Embarked'] = testX['Embarked'].dropna()
testX = pd.get_dummies(testX, columns=['Embarked'])
prediction = classifier.predict(testX)
```
